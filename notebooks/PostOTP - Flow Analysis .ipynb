```python
%matplotlib inline
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt

def plot_dataframe(df, title, file_name):
    objects = df['flow_type']
    y_pos = np.arange(len(objects))
    performance = df['count']

    fig = plt.gcf()
    
    plt.bar(y_pos, performance, align='center', alpha=0.5)
    plt.xticks(y_pos, objects)
    plt.xlabel('WorkFlow type')
    plt.ylabel('WorkFlow Count')
    plt.title(title)

    plt.show()
    plt.draw()
    
    fig.savefig(file_name + '.svg', format="svg")
    
```

```python
import itertools

# COMPLETE_DATA.info()
# ANOMALOUS_DATA.info()
# NON_ANOMALOUS_DATA.info()

flow_type_value_set = set(itertools.chain(COMPLETE_DATA['value'], 
                                          ANOMALOUS_DATA_1['value'], 
                                          ANOMALOUS_DATA_2['value'], 
                                          NON_ANOMALOUS_DATA_1['value'],
                                          NON_ANOMALOUS_DATA_2['value'],
                                          NON_ANOMALOUS_DATA_3['value']))

flow_df = pd.DataFrame(flow_type_value_set, columns=['value'])

def flow_indexes(df):
    indexes = list()
    for value in df['value']:
        index = flow_df.loc[flow_df['value'] == value].index.values[0]
        indexes.append(index)
        print('value:', value, 'index:', index)
    return indexes

COMPLETE_DATA['flow_type'] = flow_indexes(COMPLETE_DATA)
ANOMALOUS_DATA_1['flow_type'] = flow_indexes(ANOMALOUS_DATA_1)
ANOMALOUS_DATA_2['flow_type'] = flow_indexes(ANOMALOUS_DATA_2)
NON_ANOMALOUS_DATA_1['flow_type'] = flow_indexes(NON_ANOMALOUS_DATA_1)
NON_ANOMALOUS_DATA_2['flow_type'] = flow_indexes(NON_ANOMALOUS_DATA_2)
NON_ANOMALOUS_DATA_3['flow_type'] = flow_indexes(NON_ANOMALOUS_DATA_3)
    
# print(COMPLETE_DATA)
# print(ANOMALOUS_DATA)
# print(NON_ANOMALOUS_DATA_1)
# print(NON_ANOMALOUS_DATA_2)
# print(NON_ANOMALOUS_DATA_3)
```

### Filter data

In this section, we filter all data by the set of flow types presented in the anomalous and non anomalous data.

```python
flow_type_index_set = set(itertools.chain(ANOMALOUS_DATA_1['flow_type'], 
                                          ANOMALOUS_DATA_2['flow_type'],
                                          NON_ANOMALOUS_DATA_1['flow_type'],
                                          NON_ANOMALOUS_DATA_2['flow_type'],
                                          NON_ANOMALOUS_DATA_3['flow_type']))

flow_type_index_set = sorted(flow_type_index_set)

FILTERED_COMPLETE_DATA = COMPLETE_DATA[COMPLETE_DATA['flow_type'].isin(flow_type_index_set)]
FILTERED_ANOMALOUS_DATA_1 = ANOMALOUS_DATA_1[ANOMALOUS_DATA_1['flow_type'].isin(flow_type_index_set)]
FILTERED_ANOMALOUS_DATA_2 = ANOMALOUS_DATA_2[ANOMALOUS_DATA_2['flow_type'].isin(flow_type_index_set)]
FILTERED_NON_ANOMALOUS_DATA_1 = NON_ANOMALOUS_DATA_1[NON_ANOMALOUS_DATA_1['flow_type'].isin(flow_type_index_set)]
FILTERED_NON_ANOMALOUS_DATA_2 = NON_ANOMALOUS_DATA_2[NON_ANOMALOUS_DATA_2['flow_type'].isin(flow_type_index_set)]
FILTERED_NON_ANOMALOUS_DATA_3 = NON_ANOMALOUS_DATA_3[NON_ANOMALOUS_DATA_3['flow_type'].isin(flow_type_index_set)]
```
