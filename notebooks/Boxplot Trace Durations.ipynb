```python
# Boxplot properties

boxprops = dict(linestyle='-', linewidth=4, color='k')
medianprops = dict(linestyle='-', linewidth=4, color='k')
```

```python
boxplot = data.boxplot(showfliers=False, showmeans=True,
                boxprops=boxprops, medianprops=medianprops)

boxplot.set_title('Trace durations without Outliers')
boxplot.set_xlabel('Trace durations without Outliers')
boxplot.set_ylabel('Duration in Milliseconds')

boxplot
```

```python
boxplot = data.boxplot(boxprops=boxprops, medianprops=medianprops)

boxplot.set_title('Trace durations with Outliers')
boxplot.set_xlabel('Trace durations with Outliers')
boxplot.set_ylabel('Duration in Milliseconds')

boxplot
```
