```python
import matplotlib.pyplot as plt
import pandas as pd
import datetime

trace_file_data_1 = pd.read_csv('trace_file_count_2018_06_28.csv')
trace_file_data_2 = pd.read_csv('trace_file_count_2018_06_29.csv')

# print(trace_file_data_1)
# print(trace_file_data_2)

trace_file_data_1['from'] = pd.to_datetime(trace_file_data_1['from'])
trace_file_data_1['to'] = pd.to_datetime(trace_file_data_1['to'])
trace_file_data_2['from'] = pd.to_datetime(trace_file_data_2['from'])
trace_file_data_2['to'] = pd.to_datetime(trace_file_data_2['to'])

# Filter data frame 1
date_time_obj_1_1 = datetime.datetime.strptime('2018-06-27 21:00:00.000000', '%Y-%m-%d %H:%M:%S.%f')
date_time_obj_1_2 = datetime.datetime.strptime('2018-06-29 01:00:00.000000', '%Y-%m-%d %H:%M:%S.%f')
trace_file_data_1 = trace_file_data_1[(trace_file_data_1['from'] >= date_time_obj_1_1) & (trace_file_data_1['from'] <= date_time_obj_1_2)]

trace_file_data_1['day_hour'] = trace_file_data_1['from'].dt.strftime('%d %H') + ' - ' + trace_file_data_1['to'].dt.strftime('%d %H')

# Filter data frame 2
date_time_obj_2_1 = datetime.datetime.strptime('2018-06-28 19:00:00.000000', '%Y-%m-%d %H:%M:%S.%f')
date_time_obj_2_2 = datetime.datetime.strptime('2018-06-29 15:00:00.000000', '%Y-%m-%d %H:%M:%S.%f')
trace_file_data_2 = trace_file_data_2[(trace_file_data_2['from'] >= date_time_obj_2_1) & (trace_file_data_2['from'] <= date_time_obj_2_2)]

trace_file_data_2['day_hour'] = trace_file_data_2['from'].dt.strftime('%d %H') + ' - ' + trace_file_data_2['to'].dt.strftime('%d %H')

fig = plt.figure()
ax = fig.add_subplot(111)

trace_file_data_1.plot(kind='bar', x='day_hour', y='spans', ax=ax)
trace_file_data_1.plot(kind='bar', x='day_hour', y='traces', color='red', ax=ax)
ax.set(title='Trace and span count for file: traces-2018-06-28.jsonl')
ax.set(xlabel="Time (from [day hour] - to [day hour])", ylabel="Count")

plt.show()

fig.savefig("trace_file_count_2018_06_28_chart.png", bbox_inches="tight", dpi=100)

fig = plt.figure()
ax = fig.add_subplot(111)

trace_file_data_2.plot(kind='bar', x='day_hour', y='spans', ax=ax)
trace_file_data_2.plot(kind='bar', x='day_hour', y='traces', color='red', ax=ax)
ax.set(title='Trace and span count for file: traces-2018-06-29.jsonl')
ax.set(xlabel="Time (from [day hour] - to [day hour])", ylabel="Count")

plt.show()
plt.draw()
fig.savefig("trace_file_count_2018_06_29_chart.png", bbox_inches="tight", dpi=100)
```
